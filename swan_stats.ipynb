```python
import glob

%cd $output_path

def read_add_column(csv, wave_name):
    df = pd.read_csv(csv)
    df['wave'] = wave_name
    return df

files = glob.glob('stats_*.csv')

stats = pd.concat(read_add_column(f, f[6:9]) for f in files)

stats = stats.melt(id_vars=['location', 'parameter', 'reanalysis', 'wave'])

stats.columns = ['Location', 'Parameter', 'Wind dataset', 'Wave dataset', 'Stats', 'value']

stats = stats.pivot_table(values='value',
                          columns=['Wind dataset','Stats'],
                          index=['Location', 'Parameter', 'Wave dataset'])

stats = stats.sort_index(level=[0, 1], ascending=[True, False])

display(stats)
stats.to_csv('stats.csv')
```
